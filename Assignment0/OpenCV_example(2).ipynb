## Loading an image
First we load the image we wish to apply the filters to. For this session we'll make use of an example image from SciPy. 

```python
img = misc.face()
plt.imshow(img)
plt.show()
```

## Gaussian Smoothing
<p> Again smooth the image using gaussian blurring, however, this time make use of an OpenCV function. Notice how easy and how much faster the operation is compared to creating and convolving the kernel from scratch. </p>

```python
blur = cv2.GaussianBlur(img,(11,11),cv2.BORDER_DEFAULT)
plt.imshow(blur)
plt.show()
```
